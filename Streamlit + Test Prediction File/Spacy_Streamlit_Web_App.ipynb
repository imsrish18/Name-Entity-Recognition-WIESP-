```python
filepath = '/content/drive/MyDrive/AI3/WIESP2022-NER-TRAINING.jsonl'
dataset = pd.read_json(filepath, lines = True)

data = dataset[['ner_ids', 'ner_tags', 'tokens']]
data['idx'] = np.arange(0, len(data))
```

```python
data
```

```python
import sys
ner_id_list = []

for i in data['ner_tags']:
  temp = np.array(i)
  ner_id_list.append(temp)

ner_id_array = np.array(ner_id_list)
```

```python
word_list = []

for i in data['tokens']:
  temp = np.array(i)
  word_list.append(temp)

word_array = np.array(word_list)
```

```python
ner_id_list_2 = []

for i in ner_id_array:
  temp = np.insert(i, np.arange(1,len(i)), 'U')
  ner_id_list_2.append(temp)

ner_id_array_2 = np.array(ner_id_list_2)
```

```python
word_list_2 = []

for i in word_array:
  temp = np.insert(i, np.arange(1,len(i)), ' ')
  word_list_2.append(temp)

word_array_2 = np.array(word_list_2)
```

```python
word_len_list = []
for wordarray in word_array_2:
  temp_len = []
  for i in wordarray:
    temp_len.append(len(i))
  word_len_list.append(np.array(temp_len))
```

```python
ends_list = []
starts_list = []

count = 0
for i in word_len_list:
  p = np.cumsum(i)
  q = p - i

  ends_list.append(p)
  starts_list.append(q)
ends_list[0]
```

```python
# Converting list of tokens to single string docs

documents = []
for doc_list in word_array_2:
  doc = ''
  for word in doc_list:
    doc += word
  documents.append(doc)
```

```python
# creating the [start, end, entity] for each entity

entity_list = []
for i in range(len(ner_id_array_2)):
  entity_start_end = []
  for j in range(len(ner_id_array_2[i])):
    p = [int(starts_list[i][j]), int(ends_list[i][j]), ner_id_array_2[i][j]]
    start_end = np.array(p)
    entity_start_end.append(start_end)
  entity_list.append(np.array(entity_start_end))
```

```python
int_entity_list = []
for i in entity_list:
  p = i[:, 0].astype('int').reshape(len(i), 1)
  q = i[:, 1].astype('int').reshape(len(i), 1)
  ent_list = i[:, 2].reshape(len(i), 1)
  int_start_end = np.concatenate((p, q, ent_list), axis = 1, dtype = 'object')
  int_entity_list.append(int_start_end)

```

```python
# Deleting the 'O' & 'U' entities from the entity_list 

use_entity_list = []
for i in int_entity_list:
  del_pos = np.argwhere((i == 'O') | (i == 'U'))
  useful_entity = np.delete(i, del_pos[:, 0], axis = 0)
  use_entity_list.append(useful_entity)
```

```python
# Mapping the single srtring docs to respective useful entites

doc_entity_map = []
for doc, entitylist in zip(documents,use_entity_list):
  entity_dict = {}
  entity_dict['entities'] = entitylist.tolist()
  doc_entity_map.append([doc, entity_dict])
```

```python
with open('/content/drive/MyDrive/AI3/wiesp.json', 'w') as fp:
  json.dump(doc_entity_map, fp, indent = 2)
```

```python

import spacy
import json
import random
from spacy.training import Example

def load_data(file):
    with open(file, "r", encoding="utf-8") as f:
        data = json.load(f)
    return (data)

def save_data(file, data):
    with open (file, "w", encoding="utf-8") as f:
        json.dump(data, f, indent=4)

def train_spacy(data, iterations):
    TRAIN_DATA = data
    nlp = spacy.blank("en")
    if "ner" not in nlp.pipe_names:
        # ner = nlp.create_pipe("ner")
        ner = nlp.add_pipe("ner")
    for _, annotations in TRAIN_DATA:
        for ent in annotations.get("entities"):
            ner.add_label(ent[2])
    other_pipes = [pipe for pipe in nlp.pipe_names if pipe != "ner"]
    with nlp.disable_pipes(*other_pipes):
        optimizer = nlp.begin_training()
        for itn in range(iterations):
            print ("Starting iteration " + str(itn))
            random.shuffle(TRAIN_DATA)
            losses = {}
            for text, annotations in TRAIN_DATA:
                doc = nlp.make_doc(text)
                example = Example.from_dict(doc, annotations)

                nlp.update([example], sgd=optimizer, losses=losses, drop=0.2)
            print (losses)
    return (nlp)


TRAIN_DATA = load_data("/content/drive/MyDrive/AI3/wiesp.json")
nlp = train_spacy(TRAIN_DATA, 30)
```
